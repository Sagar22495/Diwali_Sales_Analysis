```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Loading & Cleaning Data

```python
df = pd.read_csv('Diwali Sales Data.csv',encoding= 'unicode_escape')
df
```

```python
df.drop(['Status','unnamed1'],axis = 1,inplace= True)
```

```python
df.dropna(inplace = True)
df
```

```python
# check if data is cleaned

df.isnull().sum()
```

```python
df.info()
```

# Analytics with modified Data

# Based on Gender

```python
ax = sns.countplot(x = 'Gender', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
    
```

```python
sales_gen = df.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
```

```python
sns.barplot(x = 'Gender',y = 'Amount',data = sales_gen)
```

From above we can see that most buyers are female and also count of purchase purchase is higher in case of female 

# Based on Age Group

```python
ax = sns.countplot(data = df, x= 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_Ag = df.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
```

```python
sns.barplot(x = 'Age Group',y = 'Amount',data = sales_Ag)
```

From above graph we can say that we have more buyers from age group 26-35

# Sales Amount Based on State

```python
sales_state = df.groupby(['State'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False).head(10)
```

```python
plt.figure(figsize = (17,8))
sns.barplot(x = 'State',y = 'Amount',data = sales_state)
```

From above figure we can say we have most of our buyers from Utter Pradesh

# Orders Based on State

```python
order_state = df.groupby(['State'],as_index = False)['Orders'].sum().sort_values(by = 'Orders',ascending = False).head(10)
```

```python
plt.figure(figsize = (17,8))
sns.barplot(x = 'State',y = 'Orders',data = order_state)
```

From above graph we can say most orders are placed by Uttar Pradesh and so on.

# Sales based on Marital Status and Gender

```python
gender_married = df.groupby(['Gender','Marital_Status'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False).head(10)
```

```python
plt.figure(figsize = (7,5))
sns.barplot(x = 'Marital_Status',y = 'Amount',data = gender_married,hue = 'Gender')
```

From above figure we can say that most buyers are married women.

# Sales based on Occupation

```python
sales_occupation = df.groupby(['Occupation'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
```

```python
plt.figure(figsize = (22,8))
sns.barplot(x = 'Occupation',y = 'Amount',data = sales_occupation)
```

From above figure we can say that most of the buyers are from IT Sector then followed by Healthcare and so no.
